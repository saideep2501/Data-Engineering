```python
spark.conf.set("Timezone","Asia/Kolkata")
```

```python
from pyspark.sql.functions import *
bronze_df=spark.table("bronze")
silver_df=bronze_df.select(
    col("name").alias("City"),
    col("main.temp").alias("Temperature"),
    col("main.humidity").alias("Humidity"),
    col("main.pressure").alias("Pressure"),
    col("sys.country").alias("Country"),
    col("coord.lat").alias("Latitude"),
    col("coord.lon").alias("Longitude")
)
silver_df=silver_df.withColumn("ingestion_Time",
                               current_timestamp()
                               )
silver_df.write.mode("overwrite").format("delta").saveAsTable("Silver_pipe")
```
